to make our functions asynchronous, we need to use async in our functions. 

what is async?

In Python, `async` is a keyword used in asynchronous programming to define functions, methods, or code blocks that can be executed asynchronously. Asynchronous programming allows you to write concurrent code that can perform multiple tasks simultaneously without blocking the execution flow.

When a function or method is defined with the `async` keyword, it becomes an asynchronous function. This means that the function can be paused and resumed during its execution to allow other tasks to run. Asynchronous functions are typically used when performing I/O-bound operations, such as network requests or file operations, that may take some time to complete.

To execute an asynchronous function, you need to use the `await` keyword. When an await statement is encountered within an asynchronous function, it pauses the execution of that function until the awaited task is completed. While waiting for the task to complete, the execution is returned to the event loop, allowing other tasks to run.

````python
# defining Commands to be used
async def start_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text('Hello! Thanks for chatting with me. I am a bot')

async def help_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(' I am a bot. Please type something so that i can respond')

async def custom_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text('This is a custom command')
````

how the bot will handle responses from the user

````python
# responses 

def handle_response(text: str) -> str:

    processed: str = text.lower()

    if 'hello' in processed:
        return 'Hey there'
    
    if 'how are you' in processed:
        return 'I am good'
    
    if 'i love python ' in processed:
        return 'Remember to subscribe!'
    
    return 'I do not understand the words coming out of your keyboard...'
````

create a function that handles the messages to send to the user

````python
async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):

    # we want to check if the message type is from a group
    message_type: str = update.message.chat.type
    text: str = update.message.text

    # debugger
    print(f'user ({update.message.chat.id}) in  {message_type}: "{text}"')

    # check whether message type is group or private
    if message_type == 'group':

        # if the message is from a group and a user responds using the bot username 
        if BOT_USERNAME in text:
            new_text: str = text.replace(BOT_USERNAME, '').strip()
            response: str = handle_response(new_text)
        else:
            return
    else:
        response: str = handle_response(text)

    # debugger
    print('Bot: ', response)
    await update.message.reply_text(response)
````

    * It handles the command to end the interaction.
    * Once executed, the bot stops and terminates its operation.

````python
async def handle_course(update: Update, context: ContextTypes.DEFAULT_TYPE):
    """
    This function handles the selection of a course by the user.
    It fetches the text from the user's input and retrieves the selected university and course responses from the user data.
    It then prints out the cutoff points for the selected course.
    """
    text = update.message.text

    # Fetch text from handle_university output
    start_response = context.user_data.get('university_response')
    print("start_response:", start_response)
    first_level_response = context.user_data.get('course_response')
    print("first_level_response:", first_level_response)

    # Retrieve cutoff points for the selected course and university
    filtered_df = df.loc[
        (df['PROGRAMME NAME'] == text.upper()) & (df['INSTITUTION NAME'] == first_level_response.upper())
    ]
    cutoffs = filtered_df[
        ['2015 CUTOFF', '2016 CUTOFF', '2017 CUTOFF', '2018 CUTOFF', '2019 CUTOFF', '2020 CUTOFF', '2021 CUTOFF']
    ]

    # Convert cutoffs to a dictionary
    cutoffs = cutoffs.to_dict(orient='list')
    print(cutoffs)

    # Prepare the output lines for cutoff points
    output_lines = []
    for year, cutoff_values in cutoffs.items():
        line = f"{year}: {', '.join(map(str, cutoff_values))}"
        output_lines.append(line)

    await update.message.reply_text(
        "\n".join(output_lines)
    )

""""this line goes inside the __name__ function"""

# Add handler to handle courses
app.add_handler(MessageHandler(filters.TEXT & ~(filters.COMMAND), handle_course))

````

So, we need to introduce the ConversationHandler. According to the Telegram Bot API, it is used to handle multi-step conversations with users.

It enables us to use more than one keyboard button function.

The ConversationHandler requires us to add two additional components: states, entry points, and fallbacks.

`States`: States represent the different stages or steps in the conversation. We define and assign unique state names to each step, such as "START", "CHOOSING_UNIVERSITY", "CHOOSING_COURSE", and so on. These states help keep track of the user's progress in the conversation.

`entry_points` and `Fallbacks`: Entry points define the starting points of the conversation, typically triggered by a specific command or user action. Fallbacks, on the other hand, handle unexpected or invalid inputs from users and provide appropriate responses or instructions. These entry points and fallbacks ensure that the conversation flows smoothly and handles various scenarios.

Each function represents a step in the conversation flow. Each function returns the next state to proceed to, indicating the flow of the conversation. The `exit` function serves as the fallback and ends the conversation.

The `entry_points` define the starting point of the conversation, which is the `/start` command handled by the `start_command` function. The `states` dictionary maps each state to the corresponding message handler function. The `fallbacks` define the fallback command, which is the `/exit` command handled by the `exit` function.
